```python
#total number of confirmed,deaths,recovered,,active around the world
```

```python
status_count = ['Confirmed', 'Deaths', 'Recovered', 'Active']
column_sums = df[status_count].sum().plot(kind='bar')
plt.xlabel("Status")
plt.ylabel("Count In Millions")
plt.title("Count of Population Status")
```

```python
#Confirmed cases in WHO region
```

```python
region=df.groupby(by='WHO Region').sum()
region.head
```

```python
#Confirmed cases in each WHO region
```

```python
region['Confirmed'].plot(kind='bar')
plt.title('Confirmed Cases Grouped By WHO Region')
plt.ylabel('Count In Million')
```

```python
#Distribution of Confirmed Cases by WHO Region
```

```python
region['Confirmed'].plot(kind='pie', autopct='%1.1f%%')
plt.title('Distribution of Confirmed Cases by WHO Region')
plt.show()
```

```python
#America have the highest number of confirmed cases by 53.6% followed by Europe(20%) and South East Asia(11.1%)
```

```python
#Deaths in each region
```

```python
colors=['blue','red','green','yellow','white','black']
region['Deaths'].plot(kind='pie', autopct='%1.1f%%',colors=colors)
plt.title('Total number of Deaths by WHO Region')
plt.show()
```

```python
#Recovered and active cases by WHO region
```

```python
region[['Recovered', 'Active']].plot(kind='bar')
plt.title('Recovered and Active Cases Grouped By WHO Region')
plt.ylabel('Count')
plt.show()
```

```python
#Countries Vs Confirmed cases
```

```python
x = df['Country/Region']  
y = df['Confirmed']
fig=plt.figure(figsize=(25,30))
plt.bar(x, y)

plt.xlabel('Country/Region')
plt.ylabel('Confirmed')
plt.title('Confirmed Vs Country/Region')

plt.show()
```

```python
#Top 10 Countries/Regions by Recovered Cases
```

```python
top_recovered = df.sort_values(by='Recovered', ascending=False)[:10]
plt.pie(top_recovered['Recovered'], labels=top_recovered['Country/Region'], autopct="%0.2f%%", pctdistance=0.8)
plt.title('Top 10 Countries/Regions by Recovered Cases')
plt.show()
```

```python
#Top countries Vs Active cases
```

```python
top_active = df.sort_values(by='Active', ascending=False)[:10]
plt.pie(top_active['Active'], labels=top_recovered['Country/Region'], autopct="%0.2f%%", pctdistance=0.8)
plt.title('Top 10 Countries/Region by Active Cases')
plt.show()
```

```python
#Top 10 Countries/Regions Vs Deaths
```

```python
top_deaths = df.sort_values(by='Deaths', ascending=False)[:10]
plt.pie(top_deaths['Deaths'], labels=top_recovered['Country/Region'], autopct="%0.2f%%", pctdistance=0.8)
plt.title('Top 10 Countries/Region by Deaths')
plt.show()
```

```python
#Top 10 Countries/Regions by Lowest Average Active Cases
```

```python
top_active = df.sort_values(by='Active', ascending=True)[:10]
plt.pie(top_active['Active'], labels=top_recovered['Country/Region'], autopct="%0.2f%%", pctdistance=0.8)
plt.title('Top 10 Countries/Region by Lowest Active Cases')
plt.show()
```
